```python
import os
# print(os.listdir("../input"))

# Any results you write to the current directory are saved as output.
import itertools
import os

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf

from sklearn.preprocessing import LabelBinarizer, LabelEncoder
from sklearn.metrics import confusion_matrix

from tensorflow import keras
layers = keras.layers
models = keras.models
```

```python
dataframe = pd.read_csv("iris.csv")
dataset = dataframe.values
dataset
```

```python
shuff = dataframe.sample(frac = 1)
```

```python
print(shuff)
```

```python
train_size = int(len(shuff) * .6)
print ("Train size: %d" % train_size)
print ("Test size: %d" % (len(shuff) - train_size))
```

```python
def train_test_split(data, train_size):
    train = data[:train_size]
    test = data[train_size:]
    return train, test
```

```python
train_data, test_data = train_test_split(shuff[['sepal.length','sepal.width','petal.length','sepal.length']], train_size)
train_text, test_text = train_test_split(shuff['variety'], train_size)
```

```python
train_data
```

```python
test_data
```

```python
encoder = LabelEncoder()
encoder.fit(train_text)
y_train = encoder.transform(train_text)
y_test = encoder.transform(test_text)
```

```python
# Converts the labels to a one-hot representation
num_classes = np.max(y_train) + 1
y_train = keras.utils.to_categorical(y_train, num_classes)
y_test = keras.utils.to_categorical(y_test, num_classes)
```

```python
print('y_train shape:', y_train.shape)
print('y_test shape:', y_test.shape)
```

```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
from tensorflow.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
```

```python
from tensorflow.keras import models
from tensorflow.keras import layers
model = models.Sequential()
model.add(layers.Dense(4, activation='relu', input_shape=(4,)))
#model.add(layers.Dropout(0.5))
model.add(layers.Dense(6, activation='relu'))
#model.add(layers.Dropout(0.5))
model.add(layers.Dense(3, activation='sigmoid'));
model.compile(optimizer='adam',
               loss='categorical_crossentropy',
               metrics=['accuracy'])
```

```python
history = model.fit(train_data, y_train,batch_size=2,epochs=10,verbose=2,validation_split=0.1)
```

```python
score = model.evaluate(test_data,y_test,batch_size=2, verbose=1)
```

```python
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```
